### Additional Medium 5

**Question:**  
Count how many times a given element appears in a list.

**Approach:**  
Use a counter variable.


```python
# Write your code here
numbers = [10, 20, 10, 30, 10, 40, 50]
target_element = 10
count = 0
for num in numbers:
    if num == target_element:
        count = count + 1
print(f"The number {target_element} appears {count} times.")
```
