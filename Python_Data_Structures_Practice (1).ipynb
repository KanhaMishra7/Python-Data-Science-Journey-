### Question 1: List (Order and Duplicates)

You are given a list of daily sales amounts:

```python
sales = [1200, 1500, 1200, 1800, 1500]
```

Tasks:
1. Print the total sales
2. Print the highest sale
3. Print how many times `1200` appears

````python
# write your code here
sales = [1200, 1500, 1200, 1800, 1500]
print(sum(sales))
print(max(sales))
print(sales.count(1200))
````

### Question 3: Set (Uniqueness)

You are given two lists of students who attended two different workshops:

```python
day1 = ["Rahul", "Aman", "Neha", "Rahul"]
day2 = ["Neha", "Aman", "Sita"]
```

Tasks:
1. Convert both lists into sets
2. Print students who attended both days
3. Print students who attended only Day 1

````python
# write your code here
day1 = ["Rahul", "Aman", "Neha", "Rahul"]
day2 = ["Neha", "Aman", "Sita"]
print(set(day1))
print(set(day2))
print(set(day1).intersection(set(day2)))
print(set(day1).difference(set(day2)))
````

### Question 4: Dictionary (Key-Value Mapping)

You are given marks of students in a subject:

```python
marks = {
    "Rahul": 85,
    "Aman": 90,
    "Neha": 78
}
```

Tasks:
1. Print marks of "Aman"
2. Add marks for "Sita" = 88
3. Update "Neha" marks to 80
4. Print all student names and their marks using a loop

````python
# write your code here
marks = {
    "Rahul": 85,
    "Aman": 90,
    "Neha": 78
}
print(marks["Aman"])
marks["Sita"] = 88
marks["Neha"] = 80
for key,value in marks.items():
    print(key,value)
````

### Question 5: Mixed Data Structures (Real-Life Scenario)

You are given a list of orders:

```python
orders = ["apple", "banana", "apple", "orange", "banana", "apple"]
```

Tasks:
1. Find all unique products
2. Create a dictionary that stores product and its count
3. Print each product and its count

````python
# write your code here
orders = ["apple", "banana", "apple", "orange", "banana", "apple"]
print(set(orders))

order_counts = {}
for i in orders:

    if i in order_counts:
        order_counts[i] += 1
    else:
        order_counts[i] = 1

print(order_counts)
print(orders.count("apple"))
print(orders.count("banana"))
print(orders.count("orange"))
````
