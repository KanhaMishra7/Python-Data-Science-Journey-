## Problem 1: Division by Zero (ZeroDivisionError)

**Problem:** Divide two numbers entered by the user.

**Approach:** Try dividing numbers and observe the error.

```python
#write your answer here
a = int(input("Enter numerator: "))
b = int(input("Enter denominator: "))
result = a / b
print("Result:", result)
```

```python
#write with exception handling(use try catch block)
try:
    a = int(input("Enter numerator: "))
    b = int(input("Enter denominator: "))
    result = a / b
    print("Result:", result)
except ZeroDivisionError:
    print("Error: Division by zero is not allowed.")
```

## Problem 2: Invalid Input (ValueError)

**Problem:** Convert user input into an integer.

**Approach:** Enter text instead of number to see the error.

```python
#write your answer here
num = int(input("Enter an integer: "))
print("You entered:", num)
```

```python
#write with exception handling(use try catch block)
try:
    num = int(input("Enter an integer: "))
    print("You entered:", num)
except ValueError:
    print("Error: Invalid input. Please enter a valid integer.")
```

## Problem 3: List Index Error (IndexError)

**Problem:** Access invalid index in a list.

```python
#write your answer here
my_list = [1, 2, 3]
print(my_list[3])
```

```python
#write with exception handling(use try catch block)
try:
    my_list = [1, 2, 3]
    print(my_list[3])
except IndexError:
    print("Error: Index out of range.")
```

## Problem 4: Dictionary Key Error (KeyError)

**Problem:** Access missing key from dictionary.

```python
#write your answer here
my_dict = {'a': 1, 'b': 2}
print(my_dict['c'])
```

```python
#write with exception handling(use try catch block)
try:
    my_dict = {'a': 1, 'b': 2}
    print(my_dict['c'])
except KeyError:
    print("Error: Key not found in the dictionary.")
```

## Problem 5: File Not Found (FileNotFoundError)

**Problem:** Try reading a file that does not exist.

```python
#write your answer here
with open("non_existent_file.txt", "r") as file:
    content = file.read()
```

```python
#write with exception handling(use try catch block)
try:
    with open("non_existent_file.txt", "r") as file:
        content = file.read()
except FileNotFoundError:
    print("Error: File not found.")
```
